```python
# ====================================
# 01. Google Drive Mount + config.yaml Load
# ====================================

from google.colab import drive, files

drive.mount('/content/drive')
```

```python
# ====================================
# 02. Config Load + Setup Paths
# ====================================

import os
import sys
import yaml

CONFIG_PATH = f"/content/config.yaml"

with open(CONFIG_PATH, 'r') as f:
    config = yaml.safe_load(f)

DRIVE_ROOT = f"/content/drive/MyDrive/{config['project_name']}"
RAW_DATA_PATH = os.path.join(DRIVE_ROOT, config['paths']['raw_data'])
PROCESSED_DATA_PATH = os.path.join(DRIVE_ROOT, config['paths']['processed_data'])

print(f"Raw Data Path: {RAW_DATA_PATH}")
print(f"Processed Data Path: {PROCESSED_DATA_PATH}")
```

```python
# ====================================
# 03. instance_usage Download
# ====================================
import urllib.request

BASE_URL = "https://storage.googleapis.com/clusterdata_2019_a"

# instance_usage: 실제 CPU/메모리 사용량 데이터
FILES = [
    "instance_usage-000000000000.json.gz",
    "instance_usage-000000000001.json.gz",
    "instance_usage-000000000002.json.gz",
]

for file in FILES:
    url = f"{BASE_URL}/{file}"
    save_path = os.path.join(RAW_DATA_PATH, file)

    if os.path.exists(save_path):
        print(f"이미 존재: {file}")
        continue

    print(f"다운로드 중: {file}...")
    urllib.request.urlretrieve(url, save_path)
    size_mb = os.path.getsize(save_path) / (1024 * 1024)
    print(f"완료: {file} ({size_mb:.2f} MB)")

print("\n다운로드 완료!")
```

```python
# ====================================
# 04. Check Downloaded Files
# ====================================

for file in os.listdir(RAW_DATA_PATH):
    if file.startswith('.'):
        continue
    filepath = os.path.join(RAW_DATA_PATH, file)
    size_mb = os.path.getsize(filepath) / (1024 * 1024)
    print(f"{file}: {size_mb:.2f} MB")
```

```python
# ====================================
# 05. Preview of Downloaded Files
# ====================================
import pandas as pd
import gzip
import json

file_path = os.path.join(RAW_DATA_PATH, "instance_usage-000000000000.json.gz")

data = []
with gzip.open(file_path, 'rt') as f:
    for i, line in enumerate(f):
        if i >= 10:
            break
        data.append(json.loads(line))

df = pd.DataFrame(data)
print("컬럼:", df.columns.tolist())
print("\n미리보기:")
df
```
